```python
import sys,os
sys.path.append(os.path.realpath('..'))

import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

from benchmarks import power
from mgc.independence import *
from mgc.sims import *
```

```python
import seaborn as sns
sns.color_palette('Set1')
sns.set(color_codes=True, style='white', context='talk', font_scale=2)
```

```python
MIN_SAMP_SIZE = 5
MAX_SAMP_SIZE = 100
STEP_SIZE = 5
POWER_REPS = 5
```

```python
simulations = [
#     linear,
#     exponential,
#     cubic,
#     joint_normal,
#     step,
#     quadratic,
#     w_shaped,
#     spiral,
#     uncorrelated_bernoulli,
#     logarithmic,
#     fourth_root,
#     sin_four_pi,
#     sin_sixteen_pi,
#     square,
#     two_parabolas,
#     circle,
#     ellipse,
#     diamond,
#     multiplicative_noise,
    multimodal_independence
]

tests = [
     CCA,
     Dcorr,
     HHG,
     Hsic,
     Kendall,
     Pearson,
     RV,
     Spearman
]
```

```python
def estimate_power(test, sim):
    est_power = []
    power_reps = []

    for i in tqdm(range(MIN_SAMP_SIZE, MAX_SAMP_SIZE, STEP_SIZE)):
        power_reps = [power(test, sim, n=i) for _ in range(POWER_REPS)]
        est_power.append(np.mean(power_reps))
    
    est_power = np.array(est_power)
    np.savetxt('../benchmarks/vs_samplesize/{}_{}.csv'.format(sim.__name__, test.__name__),
               est_power, delimiter=',')
    
    return np.array(est_power)
```

```python
est_power = [[estimate_power(test, sim) for test in tests] for sim in simulations]
```

```python
def plot_power():
    fig, ax = plt.subplots(nrows=4, ncols=5, figsize=(28,24))
    
    sim_title = [
        "Linear",
        "Exponential",
        "Cubic",
        "Joint Normal",
        "Step",
        "Quadratic",
        "W-Shaped",
        "Spiral",
        "Uncorrelated Bernoulli",
        "Logarithmic",
        "Fourth Root",
        "Sine 4\u03C0",
        "Sine 16\u03C0",
        "Square",
        "Two Parabolas",
        "Circle",
        "Ellipse",
        "Diamond",
        "Multiplicative",
        "Independence"
    ]
    
    for i, row in enumerate(ax):
        for j, col in enumerate(row):
            count = 5*i + j
            sim = simulations[count]
            
            for test in tests:
                power = np.genfromtxt('../benchmarks/vs_samplesize/{}_{}.csv'.format(sim.__name__, test.__name__),
                                      delimiter=',')
                x_axis = np.arange(MIN_SAMP_SIZE, MAX_SAMP_SIZE, STEP_SIZE)
                
                col.plot(x_axis, power, label=test.__name__)
                col.set_xticks([x_axis[0], x_axis[-1]])
                col.set_ylim(0, 1.05)
                col.set_yticks([0, 1])
                
    leg = plt.legend(bbox_to_anchor=(0.5, 0.1), bbox_transform=plt.gcf().transFigure,
                     ncol=5, loc='upper center')
    leg.get_frame().set_linewidth(0.0)
    for legobj in leg.legendHandles:
        legobj.set_linewidth(5.0)
    plt.subplots_adjust(hspace=.75)
```

```python
plot_power()
```
